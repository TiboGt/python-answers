Graph from links - Create a program that will create a graph or network from a series of links.

```python
import networkx as nx
import matplotlib.pyplot as plt

def create_graph(links):
    """Creates a graph from a series of link"""
    G = nx.Graph()
    G.add_edges_from(links)
    return G

def display_graph(G):
    """Plot graph with matplotlib."""
    plt.figure(figsize=(8, 6))
    nx.draw(G, with_labels=True, node_color='lightgreen', edge_color='gray', node_size=1000, font_size=12)
    plt.title("Generated graph from links")
    plt.show()

def main():
    #
    links = [
        ("Alice", "Bob"),
        ("Bob", "Charlie"),
        ("Alice", "David"),
        ("David", "Bob"),
        ("Eve", "Frank")
    ]
    
    print("Création du graphe...")
    G = create_graph(links)
    
    print("Affichage du graphe...")
    display_graph(G)

if __name__ == "__main__":
    main()
```
